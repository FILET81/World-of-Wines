```python
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
```

```python
# Importing cleaned dataset:
wines_dataset = pd.read_csv("wines_final.csv")

# Dropping column "Unnamed: 0":
wines_dataset = wines_dataset.drop(columns=["Unnamed: 0"])
```

```python
# Extracting lists of unique values from each requested column/option:
w_country = wines_dataset["country"].unique().tolist()
w_type = wines_dataset["wine_type"].unique().tolist()
w_monovarietal = ["yes", "no"]     # Boolean values!!!
w_noble = ["yes", "no"]     # Boolean values!!!
w_vintage = wines_dataset["vintage"].unique().tolist()
w_min_price, w_max_price = wines_dataset["price_usd"].min(), wines_dataset["price_usd"].max()
#w_price = wines_dataset["price_usd"].unique().tolist()

# Asking for a "country" out of the DataFrame:
while True:
    w_country_input = input(f"Please choose a country from the list: {', '.join(w_country)}: ").strip()
    if w_country_input in w_country:
        selected_country = w_country_input
        break
    else:
        print("Invalid choice. Please try again!")

# Asking for a "wine_type" out of the DataFrame:
while True:
    w_type_input = input(f"What type of wine would you like to find?: {', '.join(w_type)}: ").strip()
    if w_type_input in w_type:
        selected_type = w_type_input
        break
    else:
        print("Invalid choice. Please try again!")

# Asking for "monovarietal" wine (Yes/No) out of the DataFrame:
while True:
    w_monovarietal_input = input("Would you like a monovarietal wine?: ").strip().lower()
    if w_monovarietal_input in w_monovarietal:
        selected_monovarietal = w_monovarietal_input == "yes"     # Boolean values -> "True" for "yes" and "False" for "no"!!!
        break
    else:
        print("Invalid choice. Please enter 'Yes' or 'No'!")

# Asking for "noble_international" grape variety (Yes/No) out of the DataFrame:
while True:
    w_noble_input = input("Would you like a noble grape variety?: ").strip().lower()
    if w_noble_input in w_noble:
        selected_noble = w_noble_input == "yes"     # Boolean values -> "True" for "yes" and "False" for "no"!!!
        break
    else:
        print("Invalid choice. Please enter 'Yes' or 'No'!")

# Asking for a "vintage" out of the DataFrame:
while True:
    w_vintage_input = input(f"Please choose a vintage from the list: {', '.join(map(str, w_vintage))}: ").strip()
    if w_vintage_input.isdigit() and int(w_vintage_input) in w_vintage:
        selected_vintage = int(w_vintage_input)
        break
    else:
        print("Invalid choice. Please try again!")

# Asking for the maximum "price_usd" to be afforded out of the DataFrame: 
while True:
    w_price_input = input(f"What's the maximum price ($) -between {w_min_price} and {w_max_price}- you would like to pay?: ").strip()
    try:
        selected_max_price = float(w_price_input)
        if w_min_price <= selected_max_price <= w_max_price:
            break
        else:
            print(f"Invalid choice. Please enter a price between {w_min_price} and {w_max_price}!")
    except ValueError:
        print("Invalid input. Please enter a valid value!")

#print(f"\nSelected choice:\nCountry: {selected_country}\nWine Type: {selected_type}\nMonovarietal: {selected_monovarietal}\nNoble Variety: {selected_noble}\nVintage: {selected_vintage}\nMax Price ($): {selected_max_price}")

wines_dataset[(wines_dataset["country"]==selected_country)&(wines_dataset["wine_type"]==selected_type)&(wines_dataset["monovarietal"]==selected_monovarietal)&(wines_dataset["noble_international"]==selected_noble)&(wines_dataset["vintage"]==selected_vintage)&(wines_dataset["price_usd"]<selected_max_price)].sort_values(by="points", ascending=False).head(5)
```
